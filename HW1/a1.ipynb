Write the code for the following:
Consider two sets of points p 0 and p 1 describing a line in a 2-Dimensional (2D)
plane. Consider an arbitrary point p in 2D.
1. Compute the orthogonal distance between the point p and the the line
segment joining the points p 0 and p 1 .
2. Plot the line segment joining the points p 0 and p 1 , the point p and the
orthogonal distance between the point and the line segment.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def distance(p0,p1,p2):
    m = 1
    a = m
    b = -1
    c = -m*p0[0] + p0[1]
    if((p1[0] - p0[0])!=0):
        m = (p1[1]-p0[1])/(p1[0] - p0[0])
        a = m
        b = -1
        c = -m*p0[0] + p0[1]
    else:
        #x - p0[0]=0
        m = 1
        a = m
        b = 0
        c = -p0[0]
    # line eq: y - p0[1] = m*(x - p0[0])
    # m*x - y -m*p0[0] + p0[1] 
    mag = np.array([a,b])
    abmag = np.linalg.norm(mag)
    d = abs(a*p2[0] + b*p2[1] + c)/abmag
    alpha = (-1*a*(a*p2[0] + b*p2[1] + c)/abmag**2)+p2[0]
    beta = (-1*b*(a*p2[0] + b*p2[1] + c)/abmag**2)+p2[1]
    print("Orthogonal distance between the point p and the the line segment joining the points p 0 and p 1 is")
    print(d)
    return d,alpha,beta
```

```python
def plot(p0,p1,p2,alpha,beta):
    plt.plot([p0[0],p1[0]],[p0[1],p1[1]])
    plt.plot(p2[0],p2[1],marker = 'o')
    plt.plot([p2[0],alpha],[p2[1],beta])
    plt.legend(["line sement between p0,p1", "point p2","orthogonal distance"])
    plt.annotate("p0", (p0[0], p0[1]))
    plt.annotate("p1", (p1[0], p1[1]))
    plt.annotate("p2", (p2[0], p2[1]))
    plt.show()
```

```python
p0 = np.array([4,4])
p1 = np.array([4,1])
p2 = np.array([0,1])
d,alpha,beta = distance(p0,p1,p2)
plot(p0,p1,p2,alpha,beta)
```

```python
p0 = np.array([1,2])
p1 = np.array([4,2])
p2 = np.array([1,4])
d,alpha,beta = distance(p0,p1,p2)
plot(p0,p1,p2,alpha,beta)
```
