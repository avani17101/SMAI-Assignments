```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dist(p0,p1,p2): 
    px = p1[0]-p0[0]
    py = p1[1]-p0[1]
    mg = np.array([px,py])
    norm = np.linalg.norm(mg)
    u =  ((p2[0] - p0[0]) * px + (p2[1] - p0[1]) * py) / float(pow(norm,2))
    if u < 1:  
        u = 0
    if u > 1:
        u = 1
    x = p0[0] + u * px
    y = p0[1] + u * py
    dx = x - p2[0]
    dy = y - p2[1]
    dist = np.linalg.norm([dx,dy])
    print("Orthogonal distance between the point p and the the line segment joining the points p0 and p1 is",d)
    return dist,x,y
```

```python
p0 = np.array([1,2])
p1 = np.array([4,2])
p2 = np.array([1,4])
d,alpha,beta = dist(p0,p1,p2)
plot(p0,p1,p2,alpha,beta)
```

```python
def distance(p0,p1,p2):
    m = 1
    a = m
    b = -1
    c = -m*p0[0] + p0[1]
    if((p1[0] - p0[0])!=0):
        m = (p1[1]-p0[1])/(p1[0] - p0[0])
        a = m
        b = -1
        c = -m*p0[0] + p0[1]
    else:
        #x - p0[0]=0
        m = 1
        a = m
        b = 0
        c = -p0[0]
    # line eq: y - p0[1] = m*(x - p0[0])
    # m*x - y -m*p0[0] + p0[1] 
    print(a, b, c)
    mag = np.array([a,b])
    abmag = np.linalg.norm(mag)
    d = abs(a*p2[0] + b*p2[1] + c)/abmag
    alpha = (-1*a*(a*p2[0] + b*p2[1] + c)/abmag**2)+p2[0]
    beta = (-1*b*(a*p2[0] + b*p2[1] + c)/abmag**2)+p2[1]
    print("Orthogonal distance between the point p and the the line segment joining the points p 0 and p 1",d)
    return d,alpha,beta
```

```python
def plot(p0,p1,p2,alpha,beta):
    plt.plot([p0[0],p1[0]],[p0[1],p1[1]])
    plt.plot(p2[0],p2[1],marker = 'o')
    plt.plot([p2[0],alpha],[p2[1],beta])
    plt.legend(["line sement between p0,p1", "point p2","normal"])
    plt.show()
```

```python
p0 = np.array([1,2])
p1 = np.array([4,2])
p2 = np.array([1,4])
d,alpha,beta = distance(p0,p1,p2)
plot(p0,p1,p2,alpha,beta)
```
